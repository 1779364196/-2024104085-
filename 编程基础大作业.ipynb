# 数据合并（不用每次跑）

```python
#定义函数，处理数据
def merge_data(directory, data_name):
    'directory是要合并的数据的路径，data_name是要保存的文件名'
    
    txt_files = [f for f in os.listdir(directory) if f.endswith('.txt')]

    # 检查是否有TXT文件，异常处理
    if not txt_files:
        raise Exception("没有找到TXT文件。")

    # 合并TXT文件内容
    merged_data = []
    for txt_file in txt_files:
        file_path = os.path.join(directory, txt_file)
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                for line in file:
                    # 使用空格分割每一行数据
                    fields = line.strip().split()
                    # 将分割后的数据作为一个列表添加到merged_data中
                    merged_data.append(fields)
        except Exception as e:
            raise Exception(f"处理文件 {txt_file} 时发生错误: {e}")
            
    # 将合并后的数据写入CSV文件
    csv_file_path = f'{directory}/{data_name}.csv'  # 你想要保存的CSV文件路径
    try:
        with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            for row in merged_data:
                writer.writerow(row)  # 写入每一行数据
        print(f"文件已合并并保存为：{csv_file_path}")
    except Exception as e:
        raise Exception(f"写入CSV文件时发生错误: {e}")
```

```python
#按台风分割的函数
#输入值为全数据
#返回值为储存每个台风数据的list和储存台风名称的list
def data_spliting(data):
    slipt = np.where(data.iloc[:, 0] == 66666)[0]  # 提取出每个台风的表头，也就是分割行
    data_split = []   #储存分割后的台风数据
    storm_name = []   #储存台风的名字
    for i in range(len(slipt)-1):
        storm_name.append(data.iloc[slipt[i], 7])  #提取出台风的名字
        data_part = data.iloc[(slipt[i] + 1):slipt[i + 1], 2:4].copy()  #提取出台风的经纬度
        data_split.append(data_part)

    #上述循环扔掉了最后一个台风
    storm_name.append(data.iloc[slipt[-1], 7])
    data_last = data.iloc[(slipt[-1] + 1):, 2:4].copy()  #提取出台风的经纬度
    data_split.append(data_last)
    
    return((data_split, storm_name))
```

```python
#构造特征矩阵
def data_build(data, n):
    '''
    构造特征矩阵的函数,输入值为不包含66666行的dataframe，滑动步长为n,
    即用前n期的数据作为feature，第n+1期的数据作为label，以此类推，
    返回值为构建的feature矩阵与label，均为dataframe格式
    '''
    l = len(data.iloc[:,0])  #获取输入数据的行数
    data_in = data.copy()
    data_in.insert(0,"id",range(l))  #给输入数据重新插入序号
    data_iter = data.copy()
    data_x = data_in.copy()  #用于构造数据特征
    for i in range(1,n):
        data_iter = data_iter.iloc[1:l,:].copy(deep=True)  #将data_iter向下移动一行，去掉第一行，然后复制到新的data_iter。
        data_iter['id'] = range(l-i)  #更新 data_iter 中的 "id" 列，使其值从0开始，长度为 l-i
        data_x = pd.merge(data_x, data_iter, left_on="id", right_on="id", how='left')  #将data_x和更新后的data_iter 按照 "id" 列进行左连接，合并到data_x中
    data_x = data_x.drop('id', axis=1)
    data_x = data_x.iloc[0:(l-n), :]
    data_y = data_iter.iloc[1:l, :].copy(deep=True)
    data_y = data_y.drop('id', axis=1)
    data_y = data_y.iloc[0:(l-n), :]
    return((data_x, data_y))
```

```python
#构造训练集的feature和label
#训练集所有台风数据都放到一起，目的是训练出一个整体的模型
train_feature = pd.DataFrame()
train_label = pd.DataFrame()
for i in range(len(train_data_split)):
    (train_x, train_y) = data_build(train_data_split[i], 3)
    train_feature = pd.concat([train_feature, train_x])
    train_label = pd.concat([train_label, train_y])

train_label.columns = ['label1', 'label2']
```

# 定义类

```python
class TyphoonDataProcessor:
    def __init__(self, data):
        """
        初始化函数，接收全数据
        :param data: pandas DataFrame，包含所有台风数据
        """
        self.data = data  # 原始的全数据
        self.data_split = None  # 按照不同台风分割后的数据
        self.storm_name = None  # 储存台风的名字
        self.features = None  # 对每个台风构造特征
        self.labels = None  # 每个台风的label

    def data_splitting(self):
        """
        按台风分割数据，并将结果存储为属性
        """
        split_indices = np.where(self.data.iloc[:, 0] == 66666)[0]
        self.data_split = []  
        self.storm_name = []
        for i in range(len(split_indices) - 1):
            self.storm_name.append(self.data.iloc[split_indices[i], 7])
            data_part = self.data.iloc[(split_indices[i] + 1):split_indices[i + 1], 2:4].copy()
            self.data_split.append(data_part)

        #上述循环扔掉了最后一个台风
        self.storm_name.append(self.data.iloc[split_indices[-1], 7])
        data_last = self.data.iloc[(split_indices[-1] + 1):, 2:4].copy()
        self.data_split.append(data_last)

    def data_build(self, n):
        """
        构造特征矩阵的函数，输入值为不包含66666行的dataframe，滑动步长为n
        :param n: 滑动步长
        """
        if self.data_split is None:
            raise ValueError("Data has not been split. Please call data_splitting() first.")
        
        self.features = []
        self.labels = []
        for data_part in self.data_split:
            l = len(data_part.iloc[:, 0])
            data_in = data_part.copy()
            data_in.insert(0, "id", range(l))
            data_iter = data_in.copy()
            data_x = data_in.copy()
            for i in range(1, n):
                data_iter = data_iter.iloc[1:, :].copy(deep=True)
                data_iter['id'] = range(l - i)
                data_x = pd.merge(data_x, data_iter, left_on="id", right_on="id", how='left')
            data_x = data_x.drop('id', axis=1)
            data_x = data_x.iloc[0:(l - n), :]
            data_x.columns = ['feature_la1', 'feature_lo1', 'feature_la2', 'feature_lo2','feature_la3', 'feature_lo3']
            
            data_y = data_iter.iloc[1:, :].copy(deep=True)
            data_y = data_y.drop('id', axis=1)
            data_y = data_y.iloc[0:(l - n), :]
            data_y.columns = ['label_la', 'label_lo']
            self.features.append(data_x)
            self.labels.append(data_y)

    def get_features_labels(self):
        """
        获取构建的特征矩阵和标签
        :return: tuple，包含特征矩阵和标签，均为dataframe格式
        """
        if self.features is None or self.labels is None:
            raise ValueError("Features and labels have not been built. Please call data_build() with a valid step size first.")
        return pd.concat(self.features), pd.concat(self.labels)
```

```python
# 把训练数据按照台风分割，构造特征和标签
train_data = TyphoonDataProcessor(train_data)
train_data.data_splitting()
train_data.data_build(3)
train_features, train_labels = train_data.get_features_labels()  #把所有数据合并起来了
```

# 调参过程，并使用5折交叉验证

```python
# 定义模型
model = MLPRegressor(random_state=1)

# 定义参数网格
param_grid = {
    'hidden_layer_sizes': [(50, 50), (100, 50)],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant', 'adaptive'],
    'max_iter': [1000, 1500],
    'activation': ['tanh', 'relu']
}

# 创建GridSearchCV对象, 5折交叉验证
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, 
                           scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
'''
scoring='neg_mean_squared_error'用于指定使用负均方误差作为评分标准。
通过这种方式，模型的评估可以通过最大化负均方误差来实现最小化均方误差的目标。
这种设置在交叉验证和模型选择中非常常见，能够帮助选择最佳的模型参数或算法。
'''

# 执行网格搜索，在训练集上
grid_search.fit(X, Y)

# 输出最佳参数和最佳分数
print("最佳参数:", grid_search.best_params_)
print("最佳分数:", -grid_search.best_score_)  #输出负均方误差的相反数，也就是均方误差
```

```python
# 对每一个台风构造特征矩阵,并使用模型预测
columns = ['storm_name', 'length', 'MSE_lat', 'MSE_long', 'MSE_avg']
df = pd.DataFrame(index=range(len(test_data.storm_name)), columns=columns)
test_pred_list = []

for i in range(len(test_data.storm_name)):
    df.iloc[i, 0] = test_data.storm_name[i]
    df.iloc[i, 1] = len(test_data.data_split[i])
    
    #构建不同台风的特征、标签矩阵
    test_x = test_data.features[i]
    test_y = test_data.labels[i]
    
    #归一化
    test_x.columns = test_x.columns.astype(str)
    test_X = scaler.fit_transform(test_x)
    max1, min1 = scaler.data_max_, scaler.data_min_ 
    
    test_y.columns = test_y.columns.astype(str)
    test_Y = scaler.fit_transform(test_y)
    max2, min2 = scaler.data_max_, scaler.data_min_
    
    #拟合模型
    test_pred = best_model.predict(test_X)
    test_pred_lat = test_pred[:, 0] * (max2[0] -min2[0]) + min2[0]  #给数据乘上尺度映射回去
    test_pred_long = test_pred[:, 1] * (max2[1] -min2[1]) + min2[1]  #给数据乘上尺度映射回去
    test_pred_list.append(np.array(list(zip(test_pred_long, test_pred_lat))))  #先经度后纬度，方便后续画图
    
    #MSE
    prod_lat = test_Y[:,0]
    mse_lat = MSE(test_pred[:,0], prod_lat, squared = False)
    df.iloc[i, 2] = mse_lat
    prod_long = test_Y[:,1]
    mse_long = MSE(test_pred[:,0], prod_long, squared = False)
    df.iloc[i, 3] = mse_long
```
